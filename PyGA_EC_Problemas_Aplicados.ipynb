##Um engenheiro, preocupado com a segurança de sua obra, verifica a fixação de dado parafuso, que, por sua vez, é essencial para a estabilidade. Devido a isso, ele decidiu calcular as forças P e Q que agiam sobre essa peça, para constatar que o mesmo não apresentava risco à estrutura, conforme a figura abaixo. Qual a força resultante (em Newtons) que age no parafuso?



imagem_2024-12-17_172148178.png



**(BEER, Mecânica Vetorial Para Engenheiros - Estática - 9ª Edição, Pág. 24, Capítulo 2.1, Figura 1)**

#Solução:

Organizando-se os dados disponívels, temos que:


p = 40N

Q = 60N


Ângulo entre P e o eixo x = 20 graus


Ângulo entre Q e P = 25 graus


Com isso, nota-se que o ângulo de Q é igual a soma do ângulo entre P e o eixo x com 25 graus, resultando em um ângulo de 45 graus.


##Passo 1: Decomposição das forças P e Q em x e y:

Px = P . cos(20) = 40 . 0,940 = 37,588

Py = P . sen(20) = 40 . 0,342 = 13,681

Qx = Q . cos(45) = 60 . 0,707 = 42,426

Qy = Q . sen(45) = 60 . 0,707 = 42,426

##Passo 2: Soma-se as coordenadas x e y de cada vetor, para obter-se o vetor soma:

Rx = Px + Qx = 37,588 + 42,426 = 80,014

Ry = Py + Qy = 13,681 + 42,426 = 56,107

##Passo 3: Aplica-se o módulo, para assim descobrirmos a magnitide do vetor resultante (Força)

|R| = sqrt[(80,14)²+(56,107)²] = 97,725 Newtons

##Passo 4: Para descobrirmos o ângulo A do vetor resultante R em função ao eixo x, iremos aplicar a lógica do deslicamento de vetores, conforme a imagem a seguir:



imagem_2024-12-17_181224429.png



**(BEER, Mecânica Vetorial Para Engenheiros - Estática - 9ª Edição, Pág. 24, Capítulo 2.1, Figura 4)**



Como podemos perceber, ao deslocarmos o vetor Q para a ponta do vetor P, obtemos um triângulo, com os lados compostos pelos vetores R, P e Q. Como sabemos que a soma de todos os ângulos internos de um triângulo resulta em 180 graus, e que o ângulo entre P e Q é de 25 graus, subtraímos esse valor de 180, onde obtemos ângulo B, de 155 graus. Para sabermos o ângulo em A, Aplicaremos a lei dos senos, que nos diz que:

[sen(A)/cateto oposto a A] = [sen(B)/cateto oposto a B]

[sen(A)/Q] = [sen(B)/R]

Substituindo os valores, temos:

[sen(A)/60] = [sen(155)/97,73]

Onde obtemos que:

sen(A) = [60 . sen(155)/97,73]

Logo, A = 15,04 graus em relação a P.

Para obtermos o valor do ângulo de R em relação ao eixo x, somamos o valor do ângulo A com o valor do ângulo de P:

15,04 + 20 = 35,04 graus.

Logo, o vetor resultante R tem força 97.73N, e inclinação de 35,04 graus em relação ao eixo x.





```python
#Importa-se a biblioteca math para os cálculos utilizados

import math

# Dados fornecidos (percebe-se que theta de Q é igual a 25 + theta de P)

P = 40
Q = 60
theta_P = 20
theta_Q = 45


# Convertendo ângulos para radianos

theta_P_rad = math.radians(theta_P)
theta_Q_rad = math.radians(theta_Q)

# Decompondo as forças em componentes x e y

P_x = P * math.cos(theta_P_rad)
P_y = P * math.sin(theta_P_rad)

Q_x = Q * math.cos(theta_Q_rad)
Q_y = Q * math.sin(theta_Q_rad)

# Somando as componentes x e y
R_x = P_x + Q_x
R_y = P_y + Q_y

# Calculando a magnitude da força resultante
R = math.sqrt(R_x**2 + R_y**2)

# Calculando o ângulo da força resultante em relação ao eixo x
theta_R = math.degrees(math.atan2(R_y, R_x))

# Exibindo os resultados
print(P_x)
print(f"Componente x da força resultante (R_x): {R_x:.2f} N")
print(f"Componente y da força resultante (R_y): {R_y:.2f} N")
print(f"Magnitude da força resultante (R): {R:.2f} N")
print(f"Ângulo da força resultante em relação ao eixo x (theta_R): {theta_R:.2f}°")
```

##Problema 2 - Página 105

##Considere a rede de voleibol mostrada na figura. Determine o ângulo formado pelos cabos de sustentação AB e AC.



Captura de tela 2024-12-18 101017.png



**(BEER, Mecânica Vetorial Para Engenheiros - Estática - 9ª Edição, Pág. 105, Capítulo 3.3, Figura P3.39 e P3.40)**



##Solução:


Primeiro, ao estudarmos a localização dos vetores analisados no espaço tridimensional, podemos notar as seguintes características:

AB -> Tem deslocamento negativo em x, negativo em y e positivo em z

AC -> Tem deslocamento nulo em x, negativo em y e positivo em z.

Agora, para substituirmos os valores:

AB = -1.95m em x, -2.4m em y e 0.6m em z

AC = 0m em x, -2.4m em y 1.8m em z

Colocando em formato de vetores temos:

AB = (-1.95, -2.4, 0.6)

AC = (0, -2.4, 1.8)

Na sequência, para descobrir-mos o ângulo entre os vetores, aplicamos a fórmula *geométrica* de produto escalar:

|AB . AC| = |AB| . |AC| . cos(theta)


Substituindo os valores, temos:

|(-1.95, -2.4, 0.6) . (0, -2.4, 1.8)| =|(-1.95, -2.4, 0.6)| . |(0, -2.4, 1.8)| . cos(theta)

Onde obtemos:

6.84 = 3.15 . 3.00 . cos(theta)

Isolando o cosseno, temos:

cos(theta) = 6.84/(3.00 . 3.15)

cos(theta) = 0.72380952381

Portanto, theta = arccos(0.72380952381)

Então o ângulo entre AB e AC é de 43,63 graus.

```python
import numpy as np

# Definindo os vetores

AB = np.array([-1.95, -2.4, 0.6])
AC = np.array([0, -2.4, 1.8])

# Calculando o produto escalar

produto_escalar = np.dot(AB, AC)

# Calculando os módulos

modulo_AB = np.linalg.norm(AB)
modulo_AC = np.linalg.norm(AC)

# Calculando o cosseno do ângulo entre os vetores

cos_theta = produto_escalar / (modulo_AB * modulo_AC)

print(modulo_AB, modulo_AC, produto_escalar, cos_theta)
```

Passo 0 - Traçar os eixos em relação ao corpo analisado:

A partir daí, podemos começar a esboçar as análises e os cálculos pertinentes.

Passo 1 - Calcular o momento em relação a A:

Precisamos, primeiramente, decompor a força P  em x, y e z. Analisando o desenho, percebe-se que P não desloca-se em relação ao eixo x, deslica-se de forma positiva em y, e de forma negativa em z. Após, percebemos que P percorre diagonamente um cubo(ou seja, de lados iguais, e somente com ângulos retos), logo seu ângulo é de exatamente 45 graus. Portanto:

Px = 0 . cos(45) = 0

Py = |P| . sen(45) = 2P/√2 ou P√2

|Pz| = |Py| . tg(45) => |Pz| = |Py|

Sem restrição de sinais, temos: Pz = -Py, ou seja:

Px = 0

Py = P√2

Pz = -P√2

Logo, P = (0, P√2, -P√2) ou P = P√2j - P√2k


Após isso, precisamos ligar a aresta de interesse *A* até o ponto de origem *F*, de onde parte o vetor *P*. Para isso, analisamos o vetor *AF*, onde, ao olhar a figura percebemos que: *AF* tem representatividade *positiva* em x, *negativa* em y e nula em z, além de sabermos que, tanto em x quanto em y, sua decomposição tem magnitude *a*. Para expressar-mos isso em forma de vetor, temos:

*AF* = (a, -a, 0) ou *AF* = ai - aj

Passo 2 - Calcular o momento

Para fazermos o cálculo do momento de P em relação a A, fazemos:

M(AP) = *AF* x P

M(AP) = a(i - j) x P√2(j - k)

Pelo método dos versores, fazemos:

ai X P√2j  = aP√2k
ai X -P√2k = aP√2j
-aj X P√2j = 0
-aj X -P√2k = aP√2i

Logo, obtemos que M(AP) = aP√2(i + j + k)

Agora, para calcular-mos o momento em relação a *AB*, fazemos a projeção de M(AP) sobre *AB*, utilizando a seguinte equação:

M(AB) = i . M(AP)

Aplicando a lógica do produto misto, temos:

i . [(ai - aj) X (P√2j - P√2k)]

Substituindo o valor previamente obtido através do produto vetorial, temos:

i . (aP√2i + aP√2j + aP√2k)

M(AB) = (1, 0, 0) . (aP√2, aP√2, aP√2)

M(AB) = (aP√2 . 1 + aP√2 . 0 + aP√2 . 0) = aP√2

##Problema 6

##A tampa ABCD de uma caixa de armazenagem, de 0,61m x 1,0m é articulada ao longo do lado ab e mantida aberta com a corda DEC sem atrito e a um gancho em E. A partir disso, calcule a distância entre D e E e a distância entre E e C.



Captura de tela 2024-12-18 101935.png



**(BEER, Mecânica Vetorial Para Engenheiros - Estática - 9ª Edição, Pág. 106, Capítulo 3.4, Figura P3.47 e P3.48)**



Observando a figura, temos que:


O ponto D está nas coordenadas: (0.0, 0.11, ?)

O ponto E está nas coordenadas: (0.3, 0.71, 0)

O ponto C está nas coordenadas: (1.0, 0.11, ?)


Para sabermos os valores da coordenada Z, tanto para o ponto D, quanto para o ponto C, precisamos analisar que a superfície do vidro tem medidas iguais às da mesa, portanto, as arestas que os pontos D e C estão identificadas têm uma distância de 61cm do eixo origem. Portanto, para sabermos o deslocamento em Z, precisaremos realizar o cálculo de triângulo retângulo, em que:

0.11m é o cateto oposto;

0.61m é a hipotenusa;

Z é o cateto adjacente.

Aplicando a fórmula, temos que:

(0.61)² = (0.11)² + Z²

Z² = (0.61)² - (0.11)²

Z² = 0.36

Z = 0.6 metros.


Por fim, temos que:


O ponto D está nas coordenadas: (0.0, 0.11, 0.6)

O ponto E está nas coordenadas: (0.3, 0.71, 0.0)

O ponto C está nas coordenadas: (1.0, 0.11, 0.6)



Com as coordenadas determinadas, podemos calcular as distâncias:

Distância D para E:

d(DE) = sqrt[(0.3 - 0.0)² + (0.71 - 0.11)² + (0.0 - 0.6)²]

d(DE) = 0.9m

Distância E para C:

d(EC) = sqrt[(1.0 - 0.3)² + (0.11 - 0.71)² + (0.6 - 0.0)²]

d(EC) = 1.1m

Portanto, temos que a distância do ponto D para o ponto E é de 90cm, e a distância do ponto E para o ponto C é de 1.1m.


```python
import math

# Coordenadas dos pontos
D = (0.0, 0.11, 0.6)
E = (0.3, 0.71, 0.0)
C = (1.0, 0.11, 0.6)

# Função para calcular a distância entre dois pontos

def distancia(p1, p2):
    return math.sqrt((p2[0] - p1[0])**2 + (p2[1] - p1[1])**2 + (p2[2] - p1[2])**2)

# Distâncias

dist_DE = distancia(D, E)
dist_EC = distancia(E, C)

print(f"Distância de D para E: {dist_DE:.2f}")
print(f"Distância de E para C: {dist_EC:.2f}")
```
